```python
import numpy as np
import matplotlib.pyplot as plt
```

## Harmonic Signal Function

```python
def generate_harmonic_signal(A, phi, f, N, n):
    '''
    A - amplitude
    phi - phase
    f - frequency
    N - number of samples
    n - sample index
    '''
    return A * np.sin(2 * np.pi * f * n / N + phi)
```

## Polyharmonic Signal Function

```python
def generate_polyharmonic_signal(A, phi, f, N, n):
    '''
    A - array-like of amplitudes per harmonic (length K)
    phi - array-like of phases per harmonic (length K)
    f - array-like of frequencies per harmonic (length K)
    N - number of samples
    n - array of sample indices [0..N-1]
    Returns: ndarray signal of length N
    '''
    K = len(f)
    signal = np.zeros_like(n, dtype=float)
    for idx in range(K):
        signal += generate_harmonic_signal(A[idx], phi[idx], f[idx], N, n)
    return signal
```

## Auxiliary Functions

```python
def draw_signal(signal, title):
    plt.figure(figsize=(10, 6))
    plt.plot(signal)
    plt.title(title)
    plt.xlabel('Sample index')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()


def draw_multiple(signals, labels, title):
    """Plot multiple signals on a single canvas with distinct colors.

    signals: list of 1D arrays (length should be equal)
    labels: list of legend labels, same length as signals
    title: plot title
    """
    plt.figure(figsize=(10, 6))
    colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple']
    for idx, sig in enumerate(signals):
        color = colors[idx % len(colors)]
        plt.plot(sig, color=color, label=labels[idx])
    plt.title(title)
    plt.xlabel('Sample index')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.legend()
    plt.show()
```

## Task 2(A) Completion

```python
def task_2A():
    # Option 1 conditions
    A = 10
    f = 2
    N = 2048
    n = np.arange(N)
    phi_options = [0, np.pi/6, np.pi/4, np.pi/2, np.pi]

    signals = []
    labels = []
    for phi in phi_options:
        signals.append(generate_harmonic_signal(A, phi, f, N, n))
        labels.append(f"phi = {phi}")
    draw_multiple(signals, labels, "Task 2A: 5 signals on one canvas")
```
