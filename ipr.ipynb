```python
import numpy as np
import matplotlib.pyplot as plt
```

## Harmonic Signal Function

```python
def generate_harmonic_signal(A, phi, f, N, n):
    '''
    A - amplitude
    phi - phase
    f - frequency
    N - number of samples
    n - sample index
    '''
    return A * np.sin(2 * np.pi * f * n / N + phi)
```

## Polyharmonic Signal Function

```python
def generate_polyharmonic_signal(A, phi, f, N, n):
    '''
    A - amplitude
    phi - phase
    f - frequency
    N - number of samples
    n - sample index
    k - number of harmonics
    '''
    for k in range(k):
        signal += generate_harmonic_signal(A[k], phi[k], f[k], N, n)
    return signal
```

## Auxiliary Functions

```python
def draw_signal(signal, title):
    plt.figure(figsize=(10, 6))
    plt.plot(signal)
    plt.title(title)
    plt.xlabel('Sample index')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()
```

## Task 2(A) Completion

```python
def task_2A():
    # Option 1 conditions
    A = 10
    f = 2
    N = 2048
    n = np.arange(N)
    phi_options = [0, np.pi/6, np.pi/4, np.pi/2, np.pi]

    for phi in phi_options:
        signal = generate_harmonic_signal(A, phi, f, N, n)
        draw_signal(signal, f"Option 1: phi = {phi}")
```
